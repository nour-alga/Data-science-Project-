# **Clustering : Kmeans **

To be able to use our K mean algorithme we will try to find K using the ELbow methode first

```python
#let's scale our data first
scaler= StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
scores_1 = []

range_values = range(1, 20)

for i in range_values:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(scaled_data)
  scores_1.append(kmeans.inertia_)

plt.plot(scores_1, 'bx-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores WCSS')
plt.show()
```

We see when we reech betwee 7.5 and 8 - 9 it starts saturating we dn't have a huge drope so we will choose in our case K= 8

```python
kmeans = KMeans(8)
kmeans.fit(scaled_data)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [df.columns])
cluster_centers
```

```python
y_kmeans = kmeans.fit_predict(scaled_data)
y_kmeans
```

```python
# concatenate the clusters labels to our original dataframe
df_cluster = pd.concat([df, pd.DataFrame({'cluster':labels})], axis = 1)
df_cluster.head()
```

```python
# Plot the histogram of various clusters
for i in df.columns:
  plt.figure(figsize = (35, 5))
  for j in range(8):
    plt.subplot(1,8,j+1)
    cluster = df_cluster[df_cluster['cluster'] == j]
    cluster[i].hist(bins = 20)
    plt.title('{}    \nCluster {} '.format(i,j))

  plt.show()
```

lets' do a dimentionnality reduction

```python
pca = PCA(n_components = 2 )
principal_comp = pca.fit_transform(scaled_data)
principal_comp
```

```python
# Create a dataframe with the two components
pca_df = pd.DataFrame(data = principal_comp, columns =['pca1','pca2'])
pca_df.head()
```

```python
# Concatenate the clusters labels to the dataframe
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
ax = sns.scatterplot(x="pca1", y="pca2", hue = "cluster", data = pca_df, palette =['red','green','blue','pink','yellow','gray','purple', 'black'])
plt.show()
```
